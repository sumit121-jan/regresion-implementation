# Thoretical:

#  1. What does R-squared represent in a regression model?


# R-squared, or the coefficient of determination, represents the proportion of the variance in the dependent variable that can be explained by the independent variables in the model. An R-squared value of 0 means that the model explains none of the variability, while a value of 1 indicates that it explains all of the variability.

# 2. What are the assumptions of linear regression?

The assumptions of linear regression include:

Linearity: The relationship between the predictors and the outcome is linear.


Independence of errors: The residuals are independent of each other.


Homoscedasticity: The variance of errors is constant across all levels of the independent variable.

Normality of residuals: The residuals should be approximately normally distributed.


No multicollinearity: Independent variables should not be highly correlated.

# 3. What is the difference between R-squared and Adjusted R-squared?

R-squared measures the proportion of variance explained by the model, while Adjusted R-squared adjusts for the number of predictors in the model. 

This means Adjusted R-squared provides a more accurate measure of model performance, particularly when comparing models with different numbers of predictors. 

Adjusted R-squared can decrease if unnecessary predictors are added.

# 4. Why do we use Mean Squared Error (MSE)?
Mean Squared Error (MSE) is used to measure the average of the squares of the errors—that is, the average squared difference between the estimated values and the actual value.

 It provides a quantifiable measure of how well the model predicts the target variable, with larger errors having disproportionately higher weights due to squaring.

# 5. What does an Adjusted R-squared value of 0.85 indicate?
An Adjusted R-squared value of 0.85 indicates that 85% of the variance in the dependent variable is explained by the independent variables in the model, taking into account the number of predictors used. 

This suggests a good fit of the model to the data.

# 6. How do we check for normality of residuals in linear regression?
To check for normality of residuals, you can:

Use visual methods like histograms or Q-Q plots (quantile-quantile plots) to visually assess the distribution of residuals.

Conduct statistical tests such as the Shapiro-Wilk test or Kolmogorov-Smirnov test to evaluate whether the residuals deviate from normality.

# 7. What is multicollinearity, and how does it impact regression?
Multicollinearity occurs when two or more independent variables in a regression model are highly correlated. It can lead to:

Increased standard errors of the coefficient estimates, making them unstable and difficult to interpret.

Inflated t-statistics, leading to a risk of incorrectly rejecting the null hypothesis regarding the significance of predictors.

# 8. What is Mean Absolute Error (MAE)?
Mean Absolute Error (MAE) is the average of the absolute differences between predicted values and actual values. 

It provides a measure of how far predictions are from actual outcomes, treating all errors equally regardless of direction. 

It is often easier to interpret in practical terms compared to MSE.



# 9. What are the benefits of using an ML pipeline?
Benefits of using an ML pipeline include:

Improved reproducibility: Streamlined processes make it easier to replicate results.

Ease of maintenance: Clear steps in the pipeline make it easier to update or change parts of the workflow.

Automated model training and evaluation: Streamlined processes save time and reduce manual errors.

Better organization: Efficient workflow for managing data processing, feature engineering, and model training.

# 10. Why is RMSE considered more interpretable than MSE?
Root Mean Squared Error (RMSE) is considered more interpretable than Mean Squared Error (MSE) because RMSE is in the same unit as the response variable. 

This allows stakeholders to better understand the average prediction error directly in the context of the original data.

# 11. What does a high R-squared value mean?
A high R-squared value indicates that a large proportion of the variation in the dependent variable can be explained by the independent variables in the model.

 However, it does not automatically imply a good model fit or that the predictor variables are causal—it merely reflects correlation.

# 12. What happens if linear regression assumptions are violated?
If linear regression assumptions are violated, the model estimates might become inefficient and biased. 

Violations such as non-normality of residuals can lead to incorrect conclusions regarding hypothesis tests, while heteroscedasticity can affect the validity of confidence intervals and significance tests.


# 13. How can we address multicollinearity in regression?
Multicollinearity can be addressed by:

Removing highly correlated predictors or combining them.

Using dimensionality reduction techniques such as Principal Component Analysis (PCA).

Applying regularization techniques like Ridge regression that add a penalty to the model.

# 14. How can feature selection improve model performance in regression analysis?
Feature selection can enhance model performance by:

Reducing overfitting by eliminating irrelevant or redundant variables.
Simplifying the model, making it easier to interpret.

Often leading to improved predictive accuracy on unseen data by focusing on the most important predictors.

# 16. Why is MSE sensitive to outliers?
MSE is sensitive to outliers because the squaring of errors amplifies the impact of larger errors far more than smaller ones, which can skew the overall error metric.

 As a result, a few outliers can significantly increase the MSE, making it a less robust measure under such conditions.

# 17. What is the role of homoscedasticity in linear regression?
Homoscedasticity refers to the condition where the variance of the residuals is constant across all levels of the independent variable. 

It leads to valid estimates of the coefficients and ensures that the statistical tests (e.g., t-tests, F-tests) produce reliable results.

# 18. What is Root Mean Squared Error (RMSE)?
Root Mean Squared Error (RMSE) is the square root of the average squared differences between predicted and actual values. 

Like MSE, it measures prediction error, but RMSE is easier to interpret as it is in the same units as the target variable and provides a clear view of model performance.

# 19. Why is pickling considered risky?

Pickling is considered risky because it is Python-specific and can lead to compatibility issues across different Python versions or environments. There are also security concerns because unpickling data from an untrusted source can execute arbitrary code.

# 20. What alternatives exist to pickling for saving ML models?
Alternatives to pickling for saving machine learning models include:

Joblib: Optimized for saving large numpy arrays and models efficiently.

ONNX (Open Neural Network Exchange): A framework-agnostic format for representing deep learning models, allowing for portability across different platforms.

HDF5: A file format for storing large amounts of numerical data, useful for model weights and data serialization.

# 21. What is heteroscedasticity, and why is it a problem?

Heteroscedasticity refers to a situation in which the variance of the errors in a regression model is not constant across all levels of the independent variable. 

This violates the homoscedasticity assumption, leading to inefficient estimations, increased standard errors, and unreliable hypothesis tests.

# 23. What are the key components in building a machine learning pipeline?
Key components in building a machine learning pipeline typically include:

Data collection and preprocessing (cleaning, normalization).

Feature engineering and selection.

Model selection and training.

Model evaluation and validation.

Hyperparameter tuning.

Model deployment and monitoring.

# 1. Visualizing the Distribution of Errors (Residuals) Using Seaborn's "Diamonds" Dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Load the diamonds dataset
diamonds = sns.load_dataset('diamonds')

# Select features and target variable
X = diamonds[['carat', 'depth', 'table', 'price']]  # Example features
y = diamonds['price']

# Fit a linear regression model
model = sm.OLS(y, sm.add_constant(X)).fit()
residuals = model.resid

# Visualize the distribution of residuals
sns.histplot(residuals, kde=True)
plt.title('Distribution of Residuals')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.show()
```

# 2. Calculating and Printing MSE, MAE, and RMSE


```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Load the diamonds dataset
diamonds = sns.load_dataset('diamonds')

# Select features and target variable
X = diamonds[['carat', 'depth', 'table']]
y = diamonds['price']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Calculate MSE, MAE, RMSE
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = mse ** 0.5

print(f'Mean Squared Error (MSE): {mse}')
print(f'Mean Absolute Error (MAE): {mae}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
```

# 3. Checking the Assumptions of Linear Regression

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm

# Load the diamonds dataset
diamonds = sns.load_dataset('diamonds')

# Prepare data
X = diamonds[['carat', 'depth', 'table']]
y = diamonds['price']
model = sm.OLS(y, sm.add_constant(X)).fit()

# Check linearity with scatter plot
sns.pairplot(diamonds, x_vars=['carat', 'depth', 'table'], y_vars='price', height=4)
plt.suptitle('Scatter plots for linearity', y=1.02)
plt.show()

# Check homoscedasticity with residuals plot
residuals = model.resid
fitted = model.fittedvalues
plt.scatter(fitted, residuals)
plt.axhline(0, color='red', linestyle='dashed')
plt.title('Residuals vs Fitted Values')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.show()

# Check for multicollinearity with a correlation matrix
correlation_matrix = diamonds[['carat', 'depth', 'table', 'price']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

# 4. Creating a Machine Learning Pipeline with Feature Scaling 

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Load the diamonds dataset
diamonds = sns.load_dataset('diamonds')

# Prepare features and target
X = diamonds[['carat', 'depth', 'table']]
y = diamonds['price']

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Create a pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', LinearRegression())
])

# Fit the model
pipeline.fit(X_train, y_train)

# Predict
y_pred = pipeline.predict(X_test)

# Calculate MSE
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error (MSE) from pipeline: {mse}')
```

# 11. Fitting Polynomial Regression Models of Different Degrees

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Generate synthetic data
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X**2 + np.random.randn(100, 1)  # Quadratic relationship with noise

degrees = [1, 2, 3, 4]  # Degrees of polynomial regression
mse_list = []

# Fit polynomial regression for each degree
for degree in degrees:
    poly = PolynomialFeatures(degree=degree)
    X_poly = poly.fit_transform(X)
    
    model = LinearRegression()
    model.fit(X_poly, y)
    
    y_pred = model.predict(X_poly)
    mse = mean_squared_error(y, y_pred)
    mse_list.append(mse)
    
    # Plot the polynomial regression curve
    plt.scatter(X, y, color='blue', label='Data points', alpha=0.5)
    plt.scatter(X, y_pred, color='red', label='Polynomial Prediction', alpha=0.5)
    X_range = np.linspace(0, 2, 100).reshape(-1, 1)
    y_range = model.predict(poly.transform(X_range))
    plt.plot(X_range, y_range, label=f'Degree {degree}')
    plt.title(f'Polynomial Regression (Degree {degree})')
    plt.xlabel('X')
    plt.ylabel('y')
    plt.legend()
    plt.show()

# Print MSE for each degree
for degree, mse in zip(degrees, mse_list):
    print(f'Mean Squared Error for degree {degree}: {mse}')
```

# 14. Variance Inflation Factor (VIF) to Check for Multicollinearity

```python
import pandas as pd
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Generate synthetic data
np.random.seed(0)
X1 = 2 * np.random.rand(100, 1)  
X2 = 2 * X1 + np.random.randn(100, 1)  # X2 is correlated with X1
X3 = 3 * np.random.rand(100, 1)  
X4 = 2 * np.random.rand(100, 1)  
X5 = 0.5 * X1 + 0.5 * X3 + np.random.randn(100, 1)  # X5 has correlation as well

# Combine features into a DataFrame
X = pd.DataFrame(np.hstack((X1, X2, X3, X4, X5)), columns=['X1', 'X2', 'X3', 'X4', 'X5'])

# Calculate VIF
vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif_data)
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.model_selection import train_test_split

# Generate synthetic data
np.random.seed(0)
X = np.random.rand(100, 5)  # 5 features
y = 4 + 3 * X[:, 0] + 2 * X[:, 1] + np.random.randn(100)  # Linear relationship

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Fit Linear Regression
lin_model = LinearRegression()
lin_model.fit(X_train, y_train)

# Fit Ridge Regression
ridge_model = Ridge(alpha=1)  # Adjust alpha as needed for regularization
ridge_model.fit(X_train, y_train)

# Print coefficients and R-squared scores
print(f'Linear Regression Coefficients: {lin_model.coef_}')
print(f'Ridge Regression Coefficients: {ridge_model.coef_}')
print(f'Linear Regression R-squared score: {lin_model.score(X_test, y_test)}')
print(f'Ridge Regression R-squared score: {ridge_model.score(X_test, y_test)}')
```

# 24. Cross-Validation to Evaluate a Linear Regression Model

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

# Generate synthetic data
np.random.seed(0)
X = np.random.rand(100, 5)  # 5 features
y = 4 + 3 * X[:, 0] + 2 * X[:, 1] + np.random.randn(100)

# Fit the linear regression model
model = LinearRegression()

# Perform cross-validation
scores = cross_val_score(model, X, y, cv=5)  # 5-fold cross-validation

# Print cross-validated scores
print(f'Cross-validated R-squared scores: {scores}')
print(f'Mean R-squared score: {np.mean(scores)}')
```

# 25. Compare Polynomial Regression Models of Different Degrees

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Generate synthetic data
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 - 3 * X + 2 * X**2 + np.random.randn(100, 1)

degrees = [1, 2, 3, 4]
r2_scores = []

# Fit polynomial regression for each degree
for degree in degrees:
    poly = PolynomialFeatures(degree=degree)
    X_poly = poly.fit_transform(X)
    
    model = LinearRegression()
    model.fit(X_poly, y)
    
    y_pred = model.predict(X_poly)
    r2_scores.append(r2_score(y, y_pred))
    
    # Visualize the regression curve
    X_range = np.linspace(0, 2, 200).reshape(-1, 1)
    y_range = model.predict(poly.transform(X_range))
    
    plt.scatter(X, y, color='blue', label='Data Points')
    plt.plot(X_range, y_range, label=f'Degree {degree}')

plt.title('Polynomial Regression Comparison')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()

# Print R-squared scores for each degree
for degree, r2 in zip(degrees, r2_scores):
    print(f'Degree {degree} R-squared score: {r2}')
```
